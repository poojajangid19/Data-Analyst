#Convert to a NumPy array and print the total number of orders.

orders = [120, 135, 98, 150, 143, 160, 170, 129, 140, 132]

Convert to a NumPy array and print the total number of orders.

Increase orders on weekend days (index 5 and 6) by 20%, and print the updated array.

What is the new average daily orders?


```python
import numpy as np

orders = np.array([120, 135, 98, 150, 143, 160, 170, 129, 140, 132])
print(orders)

orders_sum = np.sum(orders)
print("total no of orders" , orders_sum)

orders[5] = orders[5] * 20
orders[6] = orders[6] * 20
print(orders)

orders_mean = np.mean(orders)
print(orders_mean)
```

Movie Rating Platform
Five users rated five movies. Rows represent users, columns represent movies.





ratings = [

[7, 8, 6, 5, 9],

[6, 5, 8, 7, 7],

[9, 7, 6, 8, 6],

[5, 6, 7, 7, 8],

[8, 9, 5, 6, 7]

]

Convert to NumPy array and print the average rating for each movie.

Which movie (index) received the highest average rating?

Replace all ratings below 6 with 6 (enforcing a minimum rating).


```python
import numpy as np

ratings  =np.array([[7, 8, 6, 5, 9],

                    [6, 5, 8, 7, 7],

                    [9, 7, 6, 8, 6],

                    [5, 6, 7, 7, 8],

                    [8, 9, 5, 6, 7]
                    ])

print("ratings array" , ratings)

ratings_mean = np.mean(ratings ,axis=0)
print("average rating for eeach movie" , ratings_mean)

ratings_highest = np.argmax(ratings)
print("higest rating for each movie" , ratings_highest)

ratings[ratings < 6] = 6
print("beloe 6 rating" , ratings)
```

Your expenses over 7 days across 3 categories: [food, travel, shopping].





expenses = [

[320, 150, 400],

[450, 120, 300],

[370, 180, 500],

[410, 160, 200],

[300, 140, 450],

[360, 130, 420],

[390, 170, 380]

]

Convert to NumPy array. Print the total expenses per category.

Find the day with the highest travel expense.

Extract only the shopping expenses as a 1D array.

Normalize all expenses (divide all values by 100 and round to 1 decimal).








```python
import numpy as np

monthly_expense  =np.array([

[320, 150, 400],

[450, 120, 300],

[370, 180, 500],

[410, 160, 200],

[300, 140, 450],

[360, 130, 420],

[390, 170, 380]

]
)

print("monthly expense array" , monthly_expense)

monthly_expense_mean = np.mean(monthly_expense,axis=0)
print("average monthly expense" , monthly_expense_mean)

highest_travel = np.argmax(monthly_expense[:],1)
print("highest travel expense month" , highest_travel)

round_expense = np.round(monthly_expense /100 ,1 )
print(round_expense)
```
